```python
# QUESTION 2: solve 1*x + 2*y - z = 2, 3*x + 4*z = 7, x - y = 0, y - z = 0
```

```python
# CMBAGENT

task = r"""
Let  
A = {{1, 2, -1}, {3, 0, 4}},  
b = {2, 7},  
C = {{1, -1, 0}, {0, 1, -1}}.  
Find x such that A x = b and x sits in the nullspace of C.
"""

results = cmbagent.one_shot(task,
                   max_rounds=10,
                   agent='engineer',
                   engineer_model='gpt-4.1-2025-04-14',
                   # work_dir="/mnt/d/GitHub/math_llm_benchmark",
                  )
```

```python
import time
from wolframalpha import Client
import nest_asyncio

nest_asyncio.apply()
client = Client(wolframalpha_app_id)

def get_wolframalpha_result(question: str, retries=3):
    for _ in range(retries): # reponse time too long for 1 try so doing multiple ones
        try:
            res = client.query(question)
            return next(res.results).text
        except Exception as e:
            print("Error:", e)
            time.sleep(2)  # wait and retry
    return "Failed after retries"

q1 = r"""
solve 1*x + 2*y - z = 2, 3*x + 4*z = 7, x - y = 0, y - z = 0
"""

print(get_wolframalpha_result(q1))
```

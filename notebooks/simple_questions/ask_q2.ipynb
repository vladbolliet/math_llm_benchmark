```python
# QUESTION 2: solve 1*x + 2*y - z = 2, 3*x + 4*z = 7, x - y = 0, y - z = 0 (expected result: x=y=z=1)
```

```python
# CMBAGENT

task = r"""
Let  
A = {{1, 2, -1}, {3, 0, 4}},  
b = {2, 7},  
C = {{1, -1, 0}, {0, 1, -1}}.  
Find x such that A x = b and x sits in the nullspace of C.
"""

results = cmbagent.one_shot(task,
                   max_rounds=10,
                   agent='engineer',
                   engineer_model='gpt-4.1-2025-04-14',
                  )
```

```python
# CORRECT
```

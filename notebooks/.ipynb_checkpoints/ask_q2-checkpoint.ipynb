```python
# QUESTION 2: inverse of matrix {{1,1,1},{1,e^(2*i*pi/3), e^(4*i*pi/3)},{1, e^(4*i*pi/3), e^(2*i*pi/3)}}
```

```python
# CMBAGENT

task = r"""
inverse of matrix {{1,1,1},{1,e^(2*i*pi/3), e^(4*i*pi/3)},{1, e^(4*i*pi/3), e^(2*i*pi/3)}}
"""

results = cmbagent.one_shot(task,
                   max_rounds=10,
                   agent='engineer',
                   engineer_model='gpt-4.1-2025-04-14',
                   # work_dir="/mnt/d/GitHub/math_llm_benchmark",
                  )
```

```python
import time
from wolframalpha import Client
import nest_asyncio

nest_asyncio.apply()
client = Client(wolframalpha_app_id)

def get_wolframalpha_result(question: str, retries=3):
    for _ in range(retries): # reponse time too long for 1 try so doing multiple ones
        try:
            res = client.query(question)
            return next(res.results).text
        except Exception as e:
            print("Error:", e)
            time.sleep(2)  # wait and retry
    return "Failed after retries"

q1 = r"""
inverse of matrix {{1,1,1},{1,e^(2*i*pi/3), e^(4*i*pi/3)},{1, e^(4*i*pi/3), e^(2*i*pi/3)}}
"""

print(get_wolframalpha_result(q1))
```

```python
# BOTH ARE CORRECT, JUST DIFFERENT FORMATS. TRYING HARDER PROBLEMS
```
